```python
a = np.array([2,3,4])    # a bir vektör
```

```python
b = a.reshape((1,3))     #a vektörünü 1x3 lük matrise dönüştürüyoruz
```

```python
a.ndim   # a değişkenin boyutu
```

```python
b.ndim   #b değişkenin boyutu
```

* neden tek boyutlu array i matris e dönüştürmemiz gerek? 
* fonksiyonların bizden beklediği 2 boyutlu array olabilir. bu yüzden vektör görünümlü olan array i fonksiyon için matris e dönüştürmemiz gerekebilir.

```python
a[np.newaxis,:]    #a tek boyutlu array newaxis ile 2 boyutlu olarak yazdırılır
```

```python
a[:,np.newaxis]    #2 boyutlu olarak sütun şeklinde yazdırılır
```
