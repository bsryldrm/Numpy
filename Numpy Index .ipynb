```python
list = [[10,20,30],[20,30,40],[40,50,60]]
```

```python
matrixDizi= np.array(list)
```

```python
matrixDizi
```

```python
matrixDizi[0]      #ilk satır
```

```python
matrixDizi[1][2]   #1. satır ve 2. sütun kesişiminde olan sayı
```
